# Bayesian Networks:

- A Bayesian network is a probabilistic graphical model that represents a set of variables and their conditional dependencies
- Bayesian networks are ideal for analysing an event that occurred and predicting the likelihood that any one of several variables was a contributing factor
- It is a progression from Bayesian Analysis which was based Bayes Theorem
- Then Bayes Theorum was extended to model the probabilistic relationships among many causally related variables
- The graphical structures that describe the relationships between the related variables are known as Bayesian Networks
