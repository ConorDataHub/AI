```python
state = [0,  3,  5,  7,
         1, 11, 13, 15,
         9,  4,  6,  8,
         2, 10, 12, 14]


def count_misplaced(s):
  counter = 0
  goal =[ 1,  3,  5,  7,
          9, 11, 13, 15,
          2,  4,  6,  8,
         10, 12, 14,  0]
  
  for i in range(len(s)):
    #We do not count the misplacement for tile zero/blank
    if (s[i] != 0) and (s[i] != goal[i]):
      counter+=1
  return counter

#Keeps track of the number of iterations of the A* algorithm
num_searches = 0

def print_state_as_matrix(state):
  for i in range(len(state)):
    # print(i)
    if(i>0) and ( i%4==0):
      print(" ")
    print(str(state[i])+" ", end="")

def move(pos_moves, p, s):
  print(f'move paramters are : {pos_moves}, {p}, {s}')
  #Iterates over all possible moves to the blank 
  #Continously count for the resultant heuristic (rh)
  #It keeps the state of the minimum heuristic
  st = s.copy() #Saving a copy of the states

  for i in range(len(pos_moves)):
    rh=999999 #initial value of a resultant heuristic
    dup_st = s.copy()
    temp = dup_st[p]
    dup_st[p] = dup_st[pos_moves[i]]
    dup_st[pos_moves[i]] = temp

    temp_rh = count_misplaced(dup_st)

    if(temp_rh < rh):
      rh=temp_rh
      st=dup_st.copy()
  
  return st, rh

h = count_misplaced(state)
print("___________________A* search num", num_searches,"_______________")
print("heuristic at start", h)
print("State:")
print_state_as_matrix(state)

while h>0:
  # Obtaining the position of the blank tile
  pos = int(state.index(0))
  # Increasing the search in A* in 1
  num_searches+=1
   
  if (pos==0):
    possible_moves_blank = [1,4] #The positions that the blank tile can move to
  elif (pos==1):
    possible_moves_blank = [0,2,5]
  elif (pos==2):
    possible_moves_blank = [1,3,6]
  elif (pos==3):
    possible_moves_blank = [2,7]
  elif (pos==4):
    possible_moves_blank = [1,5,8]
  elif (pos==5):
    possible_moves_blank = [1,4,6,9]
  elif (pos==6):
    possible_moves_blank = [2,5,7,10]
  elif (pos==7):
    possible_moves_blank = [3,6,11]
  elif (pos==8):
    possible_moves_blank = [4,9,12]
  elif (pos==9):
    possible_moves_blank = [5,8,10,13]
  elif (pos==10):
    possible_moves_blank = [6,9,11,14]
  elif (pos==11):
    possible_moves_blank = [7,10,15]
  elif (pos==12):
    possible_moves_blank = [8,13]
  elif (pos==13):
    possible_moves_blank = [9,12,14]
  elif (pos==14):
    possible_moves_blank = [10,13,15]
  elif (pos==15):
    possible_moves_blank = [11,14]
  
  # we are returning the current puzzle state and heuristic
  state, h = move(possible_moves_blank, pos, state)

  print("___________________A* search num", num_searches,"_______________")
  print("heuristic at start", h)
  print("State:")
  print_state_as_matrix(state)
```
